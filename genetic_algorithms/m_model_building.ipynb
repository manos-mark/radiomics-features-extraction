```python
rad_labels = pandas.read_csv(r"../dataset/radiogenomics_labels.csv")
rad_labels.head()
```

```python
dataset = pandas.read_csv('../pyradiomics/data/pyradiomics_extracted_features.csv')
dataset.head()
```

```python
patients_labels = dataset[lambda dataset: dataset.columns[1]]
patients_labels.head()
```

```python
import numpy as np
Hist_adenocarcinoma = pandas.DataFrame(columns = dataset.columns.values.tolist())
Hist_squamous = Hist_adenocarcinoma

EGFR_wildtype = Hist_adenocarcinoma
EGRF_mutant = Hist_adenocarcinoma
KRAS_wildtype = Hist_adenocarcinoma
KRAS_mutant = Hist_adenocarcinoma


for i in range(0,len(patients_labels)):

    if (rad_labels.loc[lambda rad_labels: rad_labels['Case ID'] == '%s' %(patients_labels[i][:(len(patients_labels[i]) -4)])]['Histology'].values[0] == 'Adenocarcinoma'):  #patient ID
        Hist_adenocarcinoma = Hist_adenocarcinoma.append(dataset[lambda dataset: dataset['Image'] == '%s'%patients_labels[i]])
    elif (rad_labels.loc[lambda rad_labels: rad_labels['Case ID'] == '%s' %(patients_labels[i][:(len(patients_labels[i]) -4)])]['Histology'].values[0] == 'Squamous cell carcinoma'):  #patient ID
        Hist_squamous = Hist_squamous.append(dataset[lambda dataset: dataset['Image'] == '%s'%patients_labels[i]])
        
    if (rad_labels.loc[lambda rad_labels: rad_labels['Case ID'] == '%s' %(patients_labels[i][:(len(patients_labels[i]) -4)])]['EGFR mutation status'].values[0] == 'Wildtype'):  #patient ID
        EGFR_wildtype = EGFR_wildtype.append(dataset[lambda dataset: dataset['Image'] == '%s'%patients_labels[i]])
    elif (rad_labels.loc[lambda rad_labels: rad_labels['Case ID'] == '%s' %(patients_labels[i][:(len(patients_labels[i]) -4)])]['EGFR mutation status'].values[0] == 'Mutant'):  #patient ID
        EGRF_mutant = EGRF_mutant.append(dataset[lambda dataset: dataset['Image'] == '%s'%patients_labels[i]])
    
    if (rad_labels.loc[lambda rad_labels: rad_labels['Case ID'] == '%s' %(patients_labels[i][:(len(patients_labels[i]) -4)])]['KRAS mutation status'].values[0] == 'Wildtype'):  #patient ID
        KRAS_wildtype = KRAS_wildtype.append(dataset[lambda dataset: dataset['Image'] == '%s'%patients_labels[i]])
    elif (rad_labels.loc[lambda rad_labels: rad_labels['Case ID'] == '%s' %(patients_labels[i][:(len(patients_labels[i]) -4)])]['KRAS mutation status'].values[0] == 'Mutant'):  #patient ID
        KRAS_mutant = KRAS_mutant.append(dataset[lambda dataset: dataset['Image'] == '%s'%patients_labels[i]])
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif

high_acc = 0
fts = 0
for nnn in range(1,16):
    total_before = total_after = 0
    for iii in range(0,10):
        
        #X= num_array_Hist          #uncomment the desired label to examine
        #y = np.array(target_Hist)
        #X= num_array_EGFR
        #y = np.array(target_EGFR)
        X= num_array_KRAS
        y = np.array(target_KRAS)

        # Split dataset with test 0.35, train 0.65
        X_train, X_test, y_train, y_test = train_test_split(
                X, y, stratify=y, test_size = 0.35
        )

        #plt.figure(1)
        #plt.clf()

        X_indices = np.arange(X.shape[-1])


        # Univariate feature selection with F-test for feature scoring
        # We use the default selection function to select the four
        # most significant features
        selector = SelectKBest(f_classif, k=nnn)
        selector.fit(X_train, y_train)
        scores = -np.log10(selector.pvalues_)
        scores /= scores.max()

        # Compare to the weights of an SVM
        clf = make_pipeline(MinMaxScaler(), LinearSVC(max_iter = 100000))
        clf.fit(X_train, y_train)

        total_before = total_before + clf.score(X_test, y_test)

        svm_weights = np.abs(clf[-1].coef_).sum(axis=0)
        svm_weights /= svm_weights.sum()


        clf_selected = make_pipeline(
                SelectKBest(f_classif, k=nnn), MinMaxScaler(), LinearSVC(max_iter = 100000)
        )
        clf_selected.fit(X_train, y_train)

        total_after = total_after + clf_selected.score(X_test, y_test)

        svm_weights_selected = np.abs(clf_selected[-1].coef_).sum(axis=0)
        svm_weights_selected /= svm_weights_selected.sum()

    print('Number of selected features:', nnn)
    print('Classification accuracy without selecting features: {:.3f}'
         .format(total_before/(iii + 1)))

    print('Classification accuracy after univariate feature selection: {:.3f}'
        .format(total_after/(iii + 1)))
    
    if ((total_after/(iii+1)) > high_acc):
        high_acc = (total_after/(iii+1))
        fts = selector
```

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import KFold

high_acc = 0
fts = 0
for nnn in range(1,16):
    total_before = total_after = 0
    
    #X= num_array_Hist          #uncomment the desired label to examine
    #y = np.array(target_Hist)
    #X= num_array_EGFR
    #y = np.array(target_EGFR)
    X= num_array_KRAS
    y = np.array(target_KRAS)

    n_splits = 5
    kfold = KFold(n_splits=5, random_state=None, shuffle=True)
    for train_index, test_index in kfold.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]

        X_indices = np.arange(X.shape[-1])


        # Univariate feature selection with F-test for feature scoring
        # We use the default selection function to select the four
        # most significant features
        selector = SelectKBest(f_classif, k=nnn)
        selector.fit(X_train, y_train)
        scores = -np.log10(selector.pvalues_)
        scores /= scores.max()

        # Compare to the weights of an SVM
        clf = make_pipeline(MinMaxScaler(), LinearSVC(max_iter = 100000))
        clf.fit(X_train, y_train)

        total_before = total_before + clf.score(X_test, y_test)

        svm_weights = np.abs(clf[-1].coef_).sum(axis=0)
        svm_weights /= svm_weights.sum()


        clf_selected = make_pipeline(
                SelectKBest(f_classif, k=nnn), MinMaxScaler(), LinearSVC(max_iter = 100000)
        )
        clf_selected.fit(X_train, y_train)

        total_after = total_after + clf_selected.score(X_test, y_test)

        svm_weights_selected = np.abs(clf_selected[-1].coef_).sum(axis=0)
        svm_weights_selected /= svm_weights_selected.sum()

    print('Number of selected features:', nnn)
    print('Classification accuracy without selecting features: {:.3f}'
         .format(total_before/n_splits))

    print('Classification accuracy after univariate feature selection: {:.3f}'
        .format(total_after/n_splits))
    
    if ((total_after/n_splits) > high_acc):
        high_acc = (total_after/n_splits)
        fts = selector
```
